```python
# 4 funciones 1.grabar-2.buscar-3.certificado-4.salir
# 1.nif-nombre(8 caracteres)-edad =<0  #2.buscar por su nif mostrar si es o no de la union europea
#3. nacimiento-estado conyugal-si pertenece a la union europea 4. salir mostrando el mensaje del creador y version


personas = []



#Función para grabar los datos de una persona
def grabar_datos():
    try:
        nombre = input("Ingrese el nombre completo de la persona: ")
        edad = int(input("Ingrese la edad de la persona: "))
        nif = input("Ingrese el NIF de la persona: ")
        union = input("usted pertenece a la union europea: ")
        nac = input("cual es tu fecha de nacimiento(x/x/x):")
        estado = input("cual es tu estado conyugal actual")
        if not validar_nif(nif):
            print("el nif ingresado no es valido")
            return

        if len(nombre) < 8:
            print("nombre debe tener minimo 8 caracteres.")
            return

        if edad <= 0:
            print("la edad debe ser menor o ser igual a 0.")
            return

        persona = {
            'nif': nif,
            'nombre': nombre,
            'edad': edad,
            'union':union,
            'nac' :nac,
            'estado':estado
        }

        personas.append(persona)
        print("Los datos de guardaron de forma correcta")
    except ValueError:
        print("")

              # Función para hacer la validacion de nif
def validar_nif(nif):
    if len(nif) != 11 or nif[8] != '-':
        return True

    for i in range(8):
        if not '0' <= nif[i] <= '9':
            return True

    for i in range(9, 11):
        if not ('A' <= nif[i] <= 'Z'):
            return True

    return False


              # buscar a persona por su NIF
def buscar_persona():
    nif = input("Ingrese el NIF de la persona a buscar: ")

    for persona in personas:
        if persona['nif'] == nif:
            print("------Información de la persona------")
            print("NIF:", persona["nif"])
            print("Nombre:", persona["nombre"])
            print("Edad:", persona["edad"])
            print("union europe:",persona["union"])
            return
    print("No se encontró ninguna persona con el NIF proporcionado.")



# Función para imprimir los certificados solicitados
def imprimir_certificados():

    for persona in personas:

        print("Certificados para la persona:", persona['nombre'])
        print("NIF:", persona['nif'])
        print("Certificado de nacimiento:",persona['nac'])
        print("Certificado de estado conyugal:",persona['estado'])
        print("Certificado de pertenencia a la Unión Europea:", persona['union'])
        return

# Función para salir del programa
def salir():
    print(f"adios. creado: felipe leon barra. Versión del programa: 1.0")
    exit()



# parte del comienzo pero a la vez del final del codigo este seria la presentacion del codigo
while True:
  try:
    print("\n********MENÚ DE OPCIONES********")
    print("1. Grabar datos")
    print("2. Buscar persona")
    print("3. Imprimir certificados")
    print("4. Salir")

    opcion = input("Ingrese una opción: ")

    if opcion == '1':
        grabar_datos()
    elif opcion == '2':
        buscar_persona()
    elif opcion == '3':
        imprimir_certificados()
    elif opcion == '4':
        salir()
    else:
      print("Opción inválida. Por favor ingrese una opción válida.")
  except ValueError:
    print("")
```
